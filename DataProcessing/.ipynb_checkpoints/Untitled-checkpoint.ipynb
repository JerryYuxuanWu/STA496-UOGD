```python
from sklearn.datasets import make_blobs
import pandas as pd
import hdbscan
import numpy as np
```

```python
flaring_v35 = pd.read_csv('../data/pb_vnf/flaring_v35.csv')
```

```python
flaring_v35_filtered = flaring_v35[['easting', 'northing']]
flaring_v35_filtered.shape
```

```python
# find best min cluster size by smallest noise
# try min_cluster_size from 2 to 16
options = list(range(2, 17))
noise = [0] * len(options)
clustered_obs = [0] * len(options)
clusters = [0] * len(options)
for i in range(0, len(options)):    
    clusterer = hdbscan.HDBSCAN(min_cluster_size=options[i])
    clusterer.fit(flaring_v35_filtered)
    clustered_obs[i] = np.count_nonzero(clusterer.labels_>=0)
    noise[i] = np.count_nonzero(clusterer.labels_==-1)
    clusters[i] = np.max(clusterer.labels_)
```

```python
result = pd.DataFrame({'min_pts': options, 'clustered_obs': clustered_obs, 'noise': noise, 'clusters': clusters})
result
```

```python
# Since min_pts = 4 gave the fewest noise...
clusterer = hdbscan.HDBSCAN(min_cluster_size=4)
clusterer.fit(flaring_v35_filtered)
```

```python
flaring_v35['cluster'] = clusterer.labels_
flaring_v35
```

```python
flaring_v35.to_csv('flaring_v35_clustered.csv', index=False)  
```
